```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("enhanced_zomato_dataset_clean.csv", encoding="latin1")  
```

```python
df = df.drop(columns=[
    'Restaurant_Name', 'Dining_Rating', 'Cuisine',
    'Is_Highly_Rated', 'Item_Name', 'Delivery_Rating', 'Votes'
], errors='ignore')  # errors='ignore' varsa problem yaratmaz
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['city'].value_counts().head(10).plot(kind='bar', figsize=(10,5), title='En Ã‡ok Restoran Bulunan 10 Åžehir')
plt.ylabel("Restoran SayÄ±sÄ±")
plt.show()
```

```python
df.groupby('city')['avg_rating_city'].mean().sort_values(ascending=False).head(10).plot(
    kind='bar', figsize=(10,6), color='green'
)
plt.title('Ortalama PuanÄ± En YÃ¼ksek 10 Åžehir')
plt.ylabel('Ortalama Puan')
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.show()
```

```python
print(df.columns)
```

```python
df.groupby('city')['avg_price_city'].mean().sort_values(ascending=False).head(10).plot(
    kind='bar', figsize=(10,6), color='darkred'
)
plt.title('Ortalama FiyatÄ± En YÃ¼ksek 10 Åžehir')
plt.ylabel('Fiyat')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='avg_price_restaurant', y='avg_rating_restaurant', alpha=0.6)
plt.title('Restoran FiyatÄ± ve PuanÄ± Ä°liÅŸkisi')
plt.xlabel('Ortalama Fiyat (Restoran)')
plt.ylabel('Ortalama Puan (Restoran)')
plt.ylim(0, 5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='is_expensive', y='avg_rating_restaurant')
plt.title('PahalÄ± RestoranlarÄ±n Ortalama PuanÄ±')
plt.xlabel('PahalÄ± mÄ±? (True/False)')
plt.ylabel('Ortalama Puan')
plt.ylim(0, 5)
plt.show()
```

```python
df.groupby('best_seller')['avg_rating_cuisine'].mean().sort_values(ascending=False).head(10).plot(
    kind='bar', color='orange', figsize=(10,6)
)
plt.title('En YÃ¼ksek PuanlÄ± Mutfaklar (Best Seller)')
plt.ylabel('Ortalama Puan')
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.show()
```

```python
numeric_cols = df[['avg_rating_restaurant', 'avg_price_restaurant', 'avg_price_city', 'avg_rating_city']]

plt.figure(figsize=(8,6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Korelasyon Matrisi')
plt.show()
```
